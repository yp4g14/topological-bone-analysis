```python
import skimage.filters as f
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "D:/data/OS_SHG/"
filenames = ["Male_A2_K10_SHG.tif", "Female_A3_B1_SHG.tif"]
images= []
for file in filenames:
    images.append(np.array(Image.open(path+file)))

fig, ax = plt.subplots(1,2)
for i in range(2):
    ax[i].imshow(images[i])
plt.show()
```

```python
frangi_images = []
sato_images = []
meijering_images = []
for i in range(2):
    frangi_images.append(f.frangi(images[i], black_ridges=False))
    print('frangi')
    sato_images.append(f.sato(images[i], black_ridges=False))
    print('sato')
    meijering_images.append(f.meijering(images[i], black_ridges=False))
    print('meijering')

fig, ax = plt.subplots(2,4)
for j in range(2):
    ax[j,0].imshow(images[j])
    ax[j,1].imshow(frangi_images[j], cmap='binary_r')
    ax[j,2].imshow(sato_images[j], cmap='binary_r')
    ax[j,3].imshow(meijering_images[j], cmap='binary_r')
plt.show()
```

```python
for i in range(2):
    Image.fromarray(frangi_images[i]).save(path+'preprocessing/frangi_'+filenames[i]+'.tif')
    plt.imsave(path+'preprocessing/frangi_'+filenames[i]+'2.png', frangi_images[i], cmap='binary_r')
    Image.fromarray(sato_images[i]).save(path+'preprocessing/sato_'+filenames[i]+'.tif')
    plt.imsave(path+'preprocessing/sato_'+filenames[i]+'2.png', frangi_images[i], cmap='binary_r')
    Image.fromarray(meijering_images[i]).save(path+'preprocessing/meijering_'+filenames[i]+'.tif')
    plt.imsave(path+'preprocessing/meijering_'+filenames[i]+'2.png', frangi_images[i], cmap='binary_r')
```
